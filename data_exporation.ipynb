```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
labels = ['Right_Scissors',
          'Left_Scissors',
          'Right_Needle_driver',
          'Left_Needle_driver',
          'Right_Forceps',
          'Left_Forceps',
          'Right_Empty',
          'Left_Empty']
idx_to_label = {i:inst for i, inst in zip(range(len(labels)),labels)}
class_appearance = {i:0 for i in range(len(labels))}
images_label_counts = {i:0 for i in range(len(labels))}
files_to_show = {i:[] for i in range(len(labels))}


mypath = 'datasets/labels'
images = 'datasets/images'

from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
for file in onlyfiles:
    with open(mypath+"/"+file) as f:
        lines = [line for line in f.readlines() if line !='\n']

        if len(lines) == 6:
            ofek = 5
        images_label_counts[len(lines)] += 1
        files_to_show[len(lines)].append(images+"/"+file.replace(".txt",".jpg"))
        for line in lines:
            try:
                class_appearance[int(line.split(" ")[0].strip())] += 1
            except:
                continue
from random import choice
files_to_show_list = [choice(v) for k,v in files_to_show.items() if len(v)>0]
print(class_appearance)
print(images_label_counts)

class_appearance = pd.DataFrame(class_appearance, index=[0])
class_appearance = class_appearance.rename(idx_to_label, axis=1)
plt.bar(x=class_appearance.columns, height=class_appearance.loc[0])
plt.xticks(rotation=45)
plt.show()
```

```python
from IPython.display import Image, display
for file_name in files_to_show_list:
    display(Image(filename=file_name))
```
